#### handling outliers in CNT_FAMILY_MEMBERS

```python
#handling the outliers by replacing the values of outliers with 0 percentile and 98 percentile
outliers_col = ['CNT_FAM_MEMBERS']
for i in outliers_col:
    percentiles = app[i].quantile([0.01,0.98]).values
    app[i][app[i] <= percentiles[0]] = percentiles[0]
    app[i][app[i] >= percentiles[1]] = percentiles[1]
```

# UNIVARIATE ANALYSIS

Defining a function for univariate analysis

```python
def Univariate_plot(column):
    plt.figure(figsize=(25,10))
    
    plt.subplot(1, 3, 1)
    pt=sns.countplot(column, data=app)
    plt.title('Distribution of '+column+' for all applicants', fontdict={'fontsize' : 16, 'color':'Green'})
    plt.xlabel(column)
    plt.xticks(rotation=90,fontsize = 15)
    plt.ylabel('Count of all applicants', fontsize = 15)
    for i in pt.patches:
        pt.annotate(f'{(i.get_height()/len(app)):.2%}', (i.get_x()+0.1, i.get_height()+50),rotation = 90) 
        
    plt.subplot(1, 3, 2)
    pt=sns.countplot(column, data=app_0)
    plt.title('Distribution of '+column+' for Non-Defaulters', fontdict={'fontsize' : 16, 'color':'Green'})
    plt.xlabel(column)
    plt.xticks(rotation=90,fontsize = 15)
    plt.ylabel('Count of non-defaulters',fontsize = 15)
    for i in pt.patches:
        pt.annotate(f'{(i.get_height()/len(app_0)):.2%}', (i.get_x()+0.1, i.get_height()+50),rotation = 90)

        
    plt.subplot(1, 3, 3)
    pt=sns.countplot(column, data=app_1)
    plt.title('Distribution of '+column+' for Defaulters', fontdict={'fontsize' : 16, 'color':'Green'})
    plt.xlabel(column)
    plt.xticks(rotation=90,fontsize = 15)
    plt.ylabel('Count of defaulters',fontsize = 15)
    for i in pt.patches:
        pt.annotate(f'{(i.get_height()/len(app_1)):.2%}', (i.get_x()+0.1, i.get_height()+50),rotation = 90)    

        
```
